# Tokenization

```python
#matplotlib is the most popular Python library for producing plots and other 2D data visualizations. It was originally created by John D. Hunter (JDH) and is\ now maintained by a large team of developers.
```

```python
nlp = spacy.blank("en")
text = nlp("Matplotlib is the most popular Python library for producing plots and other 2D data visualizations. It was originally created by John D. Hunter (JDH) and is\ now maintained by a large team of developers.")
for token in text:
    print (token)
```

# Tokenization of a content in Bangla language.

```python
doc = nlp("মানুষের জীবনটা পৃথিবীর নানা জীবের ইতিহাসের নানা পরিচ্ছেদের উপসংহারে, এমন একটা কথা আছে। লোকালয়ে মানুষের মধ্যে আমরা নানা জীবজন্তুর প্রচ্ছন্ন পরিচয় পেয়ে থাকি, সে কথা জানা।")
for token in doc:
    print(token)
```
